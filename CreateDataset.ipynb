```python
#https://code.tutsplus.com/tutorials/base64-encoding-and-decoding-using-python--cms-25588
import glob
import base64
import cv2
import pandas as pd 

def encodeAllImage():
    
    #load all the images
    images = [cv2.imread(file) for file in glob.glob("*.jpg")]
    imgB64 = []
    for image in images:
        code = base64.b64encode(image)
        imgB64.append(code)


    with open('oneImage.csv', 'w') as f:
        f.write("%s" % imgB64[0])
    df = pd.read_csv("oneImage.csv",delimiter=';')
    print(df.iloc[:,0])
        
    print(imgB64[0])

encodeAllImage()   
  
```
